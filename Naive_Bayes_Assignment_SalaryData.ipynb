```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
for i in salary_train.columns:
    print(i)
    if salary_train[i].dtypes == object:
        salary_train[i] = encoder.fit_transform(salary_train[i])
    else:
      pass
```

```python
encoder = LabelEncoder()
for i in salary_test.columns:
    print(i)
    if salary_test[i].dtypes == object:
        salary_test[i] = encoder.fit_transform(salary_test[i])
    else:
        pass
```

```python
train_x = salary_train[salary_train.columns[0:13]]
train_y = salary_train[salary_train.columns[13]]
test_x = salary_test[salary_test.columns[0:13]]
test_y = salary_test[salary_test.columns[13]]
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
```

```python
pred_gnb = GaussianNB().fit(train_x, train_y).predict(test_x)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(test_y, pred_gnb)
```

```python
print('Accuracy', (10759+1209)/(10759+601+2491+1209))
```

```python
pred_nmb = MultinomialNB().fit(train_x, train_y).predict(test_x)
```

```python
confusion_matrix(test_y, pred_nmb)
```
